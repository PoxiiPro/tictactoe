```python
# Nathan Yuan
# Summer Project 7/15/2022
# Purpose of this program is for a user to be able to play tic tac toe in the terminal output against an ai opponent

import random

# /////////////////////  Global variables ////////////////////////////////////////

# a list that equals 3x3 square of -
board = [" ", " ", " ",
         " ", " ", " ",
         " ", " ", " ",]

# determines whos turn it is out of the two players
currentPlayer = "X"

# player selected team
playerTeam = "X"

# variable to hold the player that won
winner = None

# varibale to check if the game is running
gameRunning = True

# ////////////////// Print starting gameboard layout /////////////////////////

#function to print each row of the board
def printBoard(board):
  print(board[0] + "  | " + board [1] + " | " + board[2])
  print("---|---|---")
  print(board[3] + "  | " + board [4] + " | " + board[5])
  print("---|---|---")
  print(board[6] + "  | " + board [7] + " | " + board[8])

# /////////// Take user input /////////////

# starting function to take user team choice and set the starting team at random 50/50 chance
def gameStart():
  global playerTeam
  playerTeam = input("Do you want to be X or O?")

  while playerTeam != "X" and playerTeam != "O":
    print("Invalid team, please pick again.")
    playerTeam = input("Do you want to be X or O?")

  global currentPlayer
  ai = 0
  player = 1
  startingPlayer = random.randint(0, 1)
  if startingPlayer == 0 and playerTeam != "X":
    currentPlayer = "X"
    print("Ai randomly selected to go first.")


  if startingPlayer == 0 and playerTeam != "O":
    currentPlayer = "O"
    print("Ai randomly selected to go first.")

  if startingPlayer == 1:
    currentPlayer = playerTeam
    print("Player randomly selected to go first.")

# function to take user input for which spot they want
def playerInput(board):
  inp = int(input("Your Turn, pick a validspot to mark, enter a number 1-9: "))  # have to convert to int since inp returns str val
  while inp < 1 or inp > 9 or board[inp - 1] != " ": # if statement to check if user inputs valid position
    print("Invalid spot... Please try again.")
    inp = int(input("Pick a valid spot to mark, enter a number not already used 1-9: "))

  if inp >= 1 and inp <= 9 and board[inp - 1] == " ": # if statement to check if user inputs valid position
    board[inp - 1] = currentPlayer
    printBoard(board)


# //////////// Check win / tie condtions, Switch the player's turn, Check win / tie condtions again loop //////////////

# function to check horizonatle win condition
def checkRow(board):
  global winner # winner var if chnaged here is changed globally

  if board[0] == board[1] == board [2] and board[1] != " ":
    winner = board[0]
    return True
  elif board[3] == board[4] == board[5] and board[3] != " ":
    winner = board[3]
    return True
  elif board[6] == board[7] == board[8] and board[6] != " ":
    winner = board[6]
    return True

# function to check verticle win condition
def checkCol(board):
  global winner

  if board[0] == board[3] == board [6] and board[0] != " ":
    winner = board[0]
    return True
  elif board[1] == board[4] == board[7] and board[1] != " ":
    winner = board[1]
    return True
  elif board[2] == board[5] == board[8] and board[2] != " ":
    winner = board[2]
    return True

# check diagnol win con
def checkDig(board):
  global winner

  if board[0] == board[4] == board [8] and board[0] != " ":
    winner = board[0]
    return True
  elif board[2] == board[4] == board[6] and board[2] != " ":
    winner = board[2]
    return True

# check for tie
def checkTie(board):
  if " " not in board:
    print("It is a tie!")
    global gameRunning
    gameRunning = False

# function to check for a win
def checkWin():
  if checkCol(board) or checkDig(board) or checkRow(board):
    print(f"The winner is {winner}!")
    global gameRunning
    gameRunning = False


# switch player
def switchPlayer():
  global currentPlayer
  if currentPlayer == "X":
    currentPlayer = "O"
  else:
    currentPlayer = "X"

# ai player, very basic - randomly picks an open spot
def ai(board):
  while currentPlayer != playerTeam:
    position = random.randint(0, 8)
    if board[position] == " ":
      if playerTeam == "X":
        board[position] = "O"
      else:
        board[position] = "X"

      switchPlayer()
      print(f"Ai's Turn... Ai picked {position + 1}")
      printBoard(board)


# ///////// actual game loop using all the functions ///////
gameStart()
printBoard(board)

while gameRunning == True:
  if currentPlayer == playerTeam:
    playerInput(board)
    checkWin()
    checkTie(board)
    switchPlayer()

  if gameRunning == False:
    break

  ai(board)
  checkWin()
  checkTie(board)
```
